```python
#pip install all the libaries that you need

!pip install pystan
!pip install fbprophet
!pip install hvplot
!pip install holoviews

#import all the libraries
import pandas as pd
import holoviews as hv
from fbprophet import Prophet
import hvplot.pandas
import datetime as dt
import numpy as np
%matplotlib inline

#here a new library we haven't used in the past. It is for making an interactive graph out of the Prophet prediction.
#https://facebook.github.io/prophet/docs/quick_start.html
from fbprophet.plot import plot_plotly, plot_components_plotly

#this is the code that will ask you to upload the csv file
from google.colab import files
uploaded = files.upload()

df_jamaica = pd.read_csv(
    "Jamaica Monthly Tourist Visitors v3.csv",
)

#create Prophet object
model_jamaica = Prophet()
#integrate dummy regressor into the model. 
#This dummy takes the value of 1 starting in March 2020 and ending (inclusive) in March 2022
model_jamaica.add_regressor('pandemic')
#fit the model to the data
model_jamaica.fit(df_jamaica)
```

```python
#the form allow the user to specify how many months to predict
number_of_predictions = 24 #@param {type:"slider", min:0, max:50, step:1}
# Create a future dataframe to hold predictions
# This dataframe should also have a column named "pandemic" 
future_jamaica_trends = model_jamaica.make_future_dataframe(periods=number_of_predictions, freq='M')

#create column called 'pandemic' and give a zero to all cells
future_jamaica_trends['pandemic']=0

#make 'ds' the index
future_jamaica_trends=future_jamaica_trends.set_index('ds')

#all cells starting in March 2020 and
#ending March 2022 should get the value of 1
future_jamaica_trends.loc['2020-03-01':'2022-03-01']['pandemic']=1

# Reset "ds" from the datetime index back to a column
future_jamaica_trends = future_jamaica_trends.reset_index()

# Make the predictions using the future_##_trends DataFrame to store the results
forecast_jamaica = model_jamaica.predict(future_jamaica_trends)

# Display the first five rows of the forecast_##_trends DataFrame
forecast_jamaica

#make interective graph using plotly
display(plot_plotly(model_jamaica, forecast_jamaica))

#make component graph using plotply
display(plot_components_plotly(model_jamaica, forecast_jamaica))
```
